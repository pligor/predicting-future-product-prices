## GRU cell - without EOS

```python
rnn_cell = PriceHistoryDummySeq2Seq.RNN_CELLS.GRU
num_epochs = 10
num_epochs, num_units, batch_size
```

```python
def experiment():
    return model.run(
        npz_path=npz_path,
        epochs=num_epochs,
        batch_size=batch_size,
        num_units=num_units,
        input_len = input_len,
        target_len = target_len,
        rnn_cell=rnn_cell,
    )
```

```python
#dyn_stats = experiment()
dyn_stats, preds_dict = get_or_run_nn(experiment, filename='005_gru_dummy_seq2seq_noEOS_60to30_prev_input',
                                     nn_runs_folder= data_folder + '/nn_runs')
```

```python
dyn_stats.plotStats()
plt.show()
```

```python
r2_scores = [r2_score(y_true=dp.targets[ind], y_pred=preds_dict[ind])
            for ind in range(len(dp.targets))]
```

```python
ind = np.argmin(r2_scores)
ind
```

```python
reals = dp.targets[ind]
preds = preds_dict[ind]
```

```python
r2_score(y_true=reals, y_pred=preds)
```

```python
sns.tsplot(data=dp.inputs[ind].flatten())
```

```python
fig = plt.figure(figsize=(15,6))
plt.plot(reals, 'b')
plt.plot(preds, 'g')
plt.legend(['reals','preds'])
plt.show()
```

```python
%%time
dtw_scores = [fastdtw(dp.targets[ind], preds_dict[ind])[0]
             for ind in range(len(dp.targets))]
```

```python
np.mean(dtw_scores)
```

```python
coint(preds, reals)
```
