# Dickey Fuller Test
http://machinelearningmastery.com/time-series-data-stationary-python/

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def dickey_fuller_print(sequence):
    result = adfuller(sequence)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))
```

```python
def dickey_fuller_test(sequence, threshold=0.05):
    result = adfuller(sequence, maxlag=maxlag)
    return result[1] < threshold
```

```python
dickey_fuller_print(seq)
```

# Custom Transformation - Log and Rolling mean

```python
plt.plot(np.log(seq).values)
```

```python
def custom_transformation(sequence, window):
    #logged_vals = np.log(sequence.values)
    logged_vals = sequence.values
    logged_seq = sequence.copy()
    logged_seq[logged_seq.index] = logged_vals
    return logged_seq - logged_seq.rolling(window=window).mean()
```

```python
window = 3
```

```python
#pd.rolling_mean(seq, window=2).plot()
seq.rolling(window=window).mean().plot()
plt.show()
```

```python
cc = custom_transformation(seq, window=window)
cc.plot()
plt.show()
```

```python
cc_valid = cc[cc==cc]
cc_valid.shape
```

```python
dickey_fuller_print(cc_valid)
```

```python
def custom_transformation_log(sequence, window):
    logged_vals = np.log(sequence.values)
    logged_seq = sequence.copy()
    logged_seq[logged_seq.index] = logged_vals
    return logged_seq - logged_seq.rolling(window=window).mean()
```

```python
cc = custom_transformation_log(seq, window=window)
cc.plot()
plt.show()
```

```python
cc_valid = cc[cc==cc]
cc_valid.shape
```

```python
dickey_fuller_print(cc_valid)
```

### Let's see how many customly transfored seqs pass the Dickey Fuller test with the high (easy) threshold

```python
def get_percentage_passed_dickey_fuller_test(window, transformation = custom_transformation, verbose=False):
    seqs_tr = [transformation(cur_seq, window=window) for cur_seq in seqs]
    if verbose:
        print len(seqs_tr)
    dicful_res_tr = [dickey_fuller_test(cur_seq[window:])
                     for cur_seq in seqs_tr if len(cur_seq) - window > maxlag + 1]
    if verbose:
        print len(dicful_res_tr)
    passed = np.sum(dicful_res_tr)
    if verbose:
        print passed
    return passed / len(dicful_res_tr)
```

```python
get_percentage_passed_dickey_fuller_test(window=3)
```

```python
windows = np.arange(2, 15)
```

```python
percents_passed = [get_percentage_passed_dickey_fuller_test(window=cur_win) for cur_win in windows]
```

```python
plt.figure(figsize=(15,6))
plt.plot(windows, percents_passed)
plt.show()
```
