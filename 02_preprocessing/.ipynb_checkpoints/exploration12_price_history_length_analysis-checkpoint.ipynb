```python
# -*- coding: UTF-8 -*-
from __future__ import division
import numpy as np
import pandas as pd
import sys
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from price_history import PriceHistory
from csv import QUOTE_ALL
```

```python
input_len = 60
target_len = 30
min_seq_len = input_len + target_len
min_seq_len
```

```python
phs = pd.read_csv('../price_history_00.csv', index_col=0, encoding='utf-8', quoting=QUOTE_ALL)
phs.shape
```

```python
mobs = pd.read_csv('../mobiles_02_no_outliers.csv', index_col=0, encoding='utf-8', quoting=QUOTE_ALL)
mobs.shape
```

```python
pricehists = phs.loc[mobs.index]
pricehists.shape
```

```python
phslong = pricehists[pricehists['sequence_length'] >= min_seq_len]
phslong.shape
```

```python
lengths = phslong['sequence_length'].astype(np.int)
```

```python
plt.figure(figsize=(17,7))
sns.distplot(lengths)
plt.xticks(range(min(lengths), max(lengths))[::20])
plt.show()
```
